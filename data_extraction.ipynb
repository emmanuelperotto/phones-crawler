# Importing Libraries

```python
import pandas
import crawler
```

# Finding all device model's info and Exporting data

```python
parquet_file = pandas.read_parquet("./files/flurry_user_event_date.parquet", columns=["deviceModel"])
```

```python
valid_devices = (parquet_file['deviceModel'] != "") & (parquet_file["deviceModel"].notnull())

device_models = parquet_file[valid_devices]["deviceModel"].unique()
device_models
```

```python
import time

result = pandas.DataFrame()

for idx, device_model in enumerate(device_models):
    time.sleep(1)
    soup = crawler.request_smartphone_info(device_model)
    print(f"Request number {idx+1}...")

    if soup:
        result = result.append(crawler.create_data_frame(soup, device_model))
        result.to_csv(f"./files/search_result.csv", index=False)
    else:
        print(f"Error when requesting info for {device_model}")
```
